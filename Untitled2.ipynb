```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings("ignore")

print("✅ Tous les modules sont bien installés !")
```

```python
np.random.seed(42)

N_COMPTES = 150
TRIMESTRES = ["2022-Q1","2022-Q2","2022-Q3","2022-Q4",
              "2023-Q1","2023-Q2","2023-Q3","2023-Q4",
              "2024-Q1","2024-Q2","2024-Q3","2024-Q4"]

SECTEURS = ["Finance","Industrie","Retail","Santé","Énergie","Tech"]

comptes = []
for i in range(1, N_COMPTES + 1):
    segment = "Grand Compte" if i <= 40 else "PME"
    secteur = np.random.choice(SECTEURS)
    
    if segment == "Grand Compte":
        revenu_base = np.random.randint(80000, 300000)
    else:
        revenu_base = np.random.randint(10000, 80000)
    
    comptes.append({
        "compte_id": f"C{i:03d}",
        "nom": f"Société_{i:03d}",
        "segment": segment,
        "secteur": secteur,
        "revenu_base": revenu_base
    })

df_comptes = pd.DataFrame(comptes)
print(df_comptes.head(10))
print(f"\n✅ {len(df_comptes)} comptes générés")
print(df_comptes["segment"].value_counts())
```

```python
rows = []

for _, compte in df_comptes.iterrows():
    revenu = compte["revenu_base"]
    
    for t_idx, trimestre in enumerate(TRIMESTRES):
        annee = int(trimestre[:4])
        q = trimestre.split("-")[1]
        
        # Saisonnalité : Q4 plus fort (renouvellements fin d'année)
        saisonnalite = {"Q1": 0.90, "Q2": 0.95, "Q3": 0.92, "Q4": 1.20}
        
        # Churn aléatoire (5% de chance de perdre un compte)
        churn = np.random.random() < 0.05
        
        if churn:
            revenu_trim = 0
        else:
            revenu_trim = revenu * saisonnalite[q] * (1 + (t_idx * 0.008))
            revenu_trim *= np.random.uniform(0.92, 1.08)
            revenu_trim = round(revenu_trim)
        
        rows.append({
            "compte_id": compte["compte_id"],
            "segment": compte["segment"],
            "secteur": compte["secteur"],
            "trimestre": trimestre,
            "annee": annee,
            "quarter": q,
            "t_idx": t_idx + 1,
            "revenu": revenu_trim
        })

df = pd.DataFrame(rows)
print(df.head(15))
print(f"\n✅ {len(df)} lignes générées ({N_COMPTES} comptes x {len(TRIMESTRES)} trimestres)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings("ignore")

# ── 1. DONNÉES ──────────────────────────────────────────────────────────────
np.random.seed(42)
N_COMPTES = 150
TRIMESTRES = ["2022-Q1","2022-Q2","2022-Q3","2022-Q4",
              "2023-Q1","2023-Q2","2023-Q3","2023-Q4",
              "2024-Q1","2024-Q2","2024-Q3","2024-Q4"]
SECTEURS = ["Finance","Industrie","Retail","Santé","Énergie","Tech"]

comptes = []
for i in range(1, N_COMPTES + 1):
    segment = "Grand Compte" if i <= 40 else "PME"
    secteur = np.random.choice(SECTEURS)
    revenu_base = np.random.randint(80000, 300000) if segment == "Grand Compte" else np.random.randint(10000, 80000)
    comptes.append({"compte_id": f"C{i:03d}", "segment": segment, "secteur": secteur, "revenu_base": revenu_base})

df_comptes = pd.DataFrame(comptes)

rows = []
for _, compte in df_comptes.iterrows():
    revenu = compte["revenu_base"]
    for t_idx, trimestre in enumerate(TRIMESTRES):
        annee = int(trimestre[:4])
        q = trimestre.split("-")[1]
        saisonnalite = {"Q1": 0.90, "Q2": 0.95, "Q3": 0.92, "Q4": 1.20}
        churn = np.random.random() < 0.05
        if churn:
            revenu_trim = 0
        else:
            revenu_trim = revenu * saisonnalite[q] * (1 + (t_idx * 0.008))
            revenu_trim *= np.random.uniform(0.92, 1.08)
            revenu_trim = round(revenu_trim)
        rows.append({"compte_id": compte["compte_id"], "segment": compte["segment"],
                     "secteur": compte["secteur"], "trimestre": trimestre,
                     "annee": annee, "quarter": q, "t_idx": t_idx + 1, "revenu": revenu_trim})

df = pd.DataFrame(rows)
print(f"✅ {len(df)} lignes générées")

# ── 2. FEATURES ML ──────────────────────────────────────────────────────────
df = df.sort_values(["compte_id", "trimestre"]).reset_index(drop=True)
df["lag1"]          = df.groupby("compte_id")["revenu"].shift(1)
df["lag2"]          = df.groupby("compte_id")["revenu"].shift(2)
df["lag4"]          = df.groupby("compte_id")["revenu"].shift(4)
df["moyenne_4t"]    = df.groupby("compte_id")["revenu"].transform(lambda x: x.shift(1).rolling(4).mean())
df["croissance_yoy"]= (df["revenu"] - df["lag4"]) / (df["lag4"] + 1)
df["quarter_num"]   = df["quarter"].str.replace("Q","").astype(int)
df["segment_num"]   = (df["segment"] == "Grand Compte").astype(int)
df_ml = df.dropna().reset_index(drop=True)
print(f"✅ Dataset ML : {len(df_ml)} lignes")

# ── 3. MODÈLES ──────────────────────────────────────────────────────────────
features = ["t_idx", "quarter_num", "segment_num",
            "lag1", "lag2", "lag4", "moyenne_4t", "croissance_yoy"]

train = df_ml[df_ml["annee"] < 2024]
test  = df_ml[df_ml["annee"] == 2024]
X_train, y_train = train[features], train["revenu"]
X_test,  y_test  = test[features],  test["revenu"]

rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)
mae_rf  = mean_absolute_error(y_test, pred_rf)

xgb = XGBRegressor(n_estimators=200, learning_rate=0.05, random_state=42)
xgb.fit(X_train, y_train)
pred_xgb = xgb.predict(X_test)
mae_xgb  = mean_absolute_error(y_test, pred_xgb)

print(f"\n🌲 Random Forest — MAE : {round(mae_rf):,} €")
print(f"⚡ XGBoost       — MAE : {round(mae_xgb):,} €")

# ── 4. PRÉVISIONS 2025 ──────────────────────────────────────────────────────
TRIMESTRES_2025 = ["2025-Q1","2025-Q2","2025-Q3","2025-Q4"]
previsions_2025 = []

for compte_id in df_ml["compte_id"].unique():
    df_compte = df_ml[df_ml["compte_id"] == compte_id].copy()
    segment_num = df_compte["segment_num"].iloc[0]
    revenus_connus = list(df_compte["revenu"].values)

    for q_idx, trimestre in enumerate(TRIMESTRES_2025):
        quarter_num = q_idx + 1
        t_idx       = 13 + q_idx
        lag1        = revenus_connus[-1]
        lag2        = revenus_connus[-2]
        lag4        = revenus_connus[-4]
        moyenne4t   = np.mean(revenus_connus[-4:])
        croissance_yoy = (lag1 - lag4) / (lag4 + 1)

        X_pred = pd.DataFrame([[t_idx, quarter_num, segment_num,
                                 lag1, lag2, lag4, moyenne4t, croissance_yoy]],
                               columns=features)

        prev_rf  = rf.predict(X_pred)[0]
        prev_xgb = xgb.predict(X_pred)[0]

        previsions_2025.append({
            "compte_id"   : compte_id,
            "segment"     : df_compte["segment"].iloc[0],
            "secteur"     : df_compte["secteur"].iloc[0],
            "trimestre"   : trimestre,
            "prev_rf"     : round(prev_rf),
            "prev_xgb"    : round(prev_xgb),
            "prev_moyenne": round((prev_rf + prev_xgb) / 2)
        })
        revenus_connus.append(prev_rf)

df_prev = pd.DataFrame(previsions_2025)

print("\n📊 Prévisions 2025 — Portefeuille total :\n")
resume = df_prev.groupby("trimestre")[["prev_rf","prev_xgb","prev_moyenne"]].sum() / 1_000_000
resume.columns = ["Random Forest (M€)","XGBoost (M€)","Moyenne (M€)"]
print(resume.round(2))

print("\n🏆 Top 10 comptes prévus en 2025 :")
top10_2025 = df_prev.groupby("compte_id")["prev_moyenne"].sum().sort_values(ascending=False).head(10)
print(top10_2025.apply(lambda x: f"{x/1e3:.0f} K€"))
```

```python
from openpyxl.styles import PatternFill, Font, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import openpyxl

# ── 1. Préparer les données ──────────────────────────────────────────────────

# Tous les comptes avec détail trimestriel
df_prev_detail = df_prev.merge(
    df_analyse[["compte_id","revenu_2024","croissance"]], 
    on="compte_id"
)

# Pivot trimestriel par compte
df_pivot = df_prev.pivot_table(
    index=["compte_id","segment","secteur"],
    columns="trimestre",
    values="prev_moyenne"
).reset_index()
df_pivot["Total 2025"] = df_pivot[TRIMESTRES_2025].sum(axis=1)
df_pivot = df_pivot.merge(df_analyse[["compte_id","revenu_2024","croissance"]], on="compte_id")
df_pivot = df_pivot.sort_values("Total 2025", ascending=False).reset_index(drop=True)

# ── 2. Créer le fichier Excel ─────────────────────────────────────────────────
writer = pd.ExcelWriter("previsions_2025.xlsx", engine="openpyxl")

# --- Onglet 1 : Tous les comptes ---
df_pivot.to_excel(writer, sheet_name="Tous les comptes", index=False)

# --- Onglet 2 : Grands Comptes ---
df_gc = df_pivot[df_pivot["segment"] == "Grand Compte"]
df_gc.to_excel(writer, sheet_name="Grands Comptes", index=False)

# --- Onglet 3 : PME ---
df_pme = df_pivot[df_pivot["segment"] == "PME"]
df_pme.to_excel(writer, sheet_name="PME", index=False)

# --- Onglet 4 : Comptes à risque ---
df_risque = df_analyse[df_analyse["croissance"] < 0].sort_values("croissance")
df_risque.to_excel(writer, sheet_name="Comptes à risque", index=False)

# --- Onglet 5 : Comptes à potentiel ---
df_potentiel = df_analyse[df_analyse["croissance"] > 0].sort_values("croissance", ascending=False)
df_potentiel.to_excel(writer, sheet_name="Comptes à potentiel", index=False)

# --- Onglet 6 : Résumé par secteur ---
df_secteur = df_prev.groupby(["secteur","trimestre"])["prev_moyenne"].sum().unstack().reset_index()
df_secteur["Total 2025"] = df_secteur[TRIMESTRES_2025].sum(axis=1)
df_secteur.to_excel(writer, sheet_name="Par secteur", index=False)

# --- Onglet 7 : Résumé global ---
df_global = pd.DataFrame({
    "Trimestre": TRIMESTRES_2025,
    "Random Forest (€)": df_prev.groupby("trimestre")["prev_rf"].sum().values,
    "XGBoost (€)"      : df_prev.groupby("trimestre")["prev_xgb"].sum().values,
    "Moyenne (€)"      : df_prev.groupby("trimestre")["prev_moyenne"].sum().values,
})
df_global.loc[len(df_global)] = ["TOTAL", 
    df_global["Random Forest (€)"].sum(),
    df_global["XGBoost (€)"].sum(),
    df_global["Moyenne (€)"].sum()]
df_global.to_excel(writer, sheet_name="Résumé global", index=False)

writer.close()

# ── 3. Mise en forme ─────────────────────────────────────────────────────────
wb = openpyxl.load_workbook("previsions_2025.xlsx")

# Couleurs
BLEU      = PatternFill("solid", fgColor="1F4E79")
BLEU_CLAIR= PatternFill("solid", fgColor="BDD7EE")
ROUGE     = PatternFill("solid", fgColor="FF0000")
VERT      = PatternFill("solid", fgColor="00B050")
ORANGE    = PatternFill("solid", fgColor="FFC000")
GRIS      = PatternFill("solid", fgColor="F2F2F2")

def formater_feuille(ws, couleur_header=BLEU):
    # Header
    for cell in ws[1]:
        cell.fill      = couleur_header
        cell.font      = Font(bold=True, color="FFFFFF", size=10)
        cell.alignment = Alignment(horizontal="center", vertical="center")
    
    # Largeur colonnes
    for col in ws.columns:
        max_len = max(len(str(cell.value or "")) for cell in col)
        ws.column_dimensions[get_column_letter(col[0].column)].width = min(max_len + 4, 25)
    
    # Alternance de couleurs
    for i, row in enumerate(ws.iter_rows(min_row=2), start=2):
        fill = GRIS if i % 2 == 0 else PatternFill("solid", fgColor="FFFFFF")
        for cell in row:
            cell.fill      = fill
            cell.alignment = Alignment(horizontal="center")
            # Formater les nombres
            if isinstance(cell.value, (int, float)):
                cell.number_format = "#,##0"

# Appliquer la mise en forme à chaque feuille
for sheet_name in wb.sheetnames:
    formater_feuille(wb[sheet_name])

# Colorer les croissances négatives en rouge et positives en vert
for sheet_name in ["Comptes à risque", "Comptes à potentiel", "Tous les comptes"]:
    ws = wb[sheet_name]
    for row in ws.iter_rows(min_row=2):
        for cell in row:
            if cell.column_letter == get_column_letter(ws.max_column):  # dernière colonne = croissance
                if isinstance(cell.value, (int, float)):
                    cell.fill = ROUGE if cell.value < 0 else VERT
                    cell.font = Font(bold=True, color="FFFFFF")

wb.save("previsions_2025.xlsx")
print("✅ Fichier Excel créé : previsions_2025.xlsx")
print(f"   📋 {len(wb.sheetnames)} onglets : {', '.join(wb.sheetnames)}")
```
